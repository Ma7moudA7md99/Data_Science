```python
# main code
from tkinter import *
from tkinter import ttk
main = Tk()
main.geometry('315x415+500+200')
main.resizable(False, False)
main.title("Calculator")
# ----- backgrond -----
main.configure(bg='#f4ebd0')
# ----------
# ----- function -----
my_input = StringVar()
operator = ""
def click_btns(numbers):
    global operator
    operator += (str(numbers))
    my_input.set(operator)
    
def resule_btn():
    global operator
    result = str(eval(operator))
    my_input.set(result)
    operator = result
    
def clear_btn():
    global operator
    operator = ""
    my_input.set("")
# ----------
# ----- entry -----
title = Entry(main, text="Mahmoud", font=('PT Heading', 28, 'bold'),
              textvariable= my_input,
                      bg='#b68d40',
                      fg='white',
                      width=14)
title.place(x=8, y=10)
# ----------
# ----- BTN's -----
btn0 = Button(main, text='0', width=4, height= 2,
             font=('PT Heading', 18, 'bold'),
                      bg='#b68d40',
                      fg='white',
             command = lambda:click_btns(0))
btn0.place(x=8, y= 323)
btn1 = Button(main, text='1', width=4, height= 2,
             font=('PT Heading', 18, 'bold'),
                      bg='#b68d40',
                      fg='white',
             command = lambda:click_btns(1))
btn1.place(x=8, y= 237)
btn2 = Button(main, text='2', width=4, height= 2,
             font=('PT Heading', 18, 'bold'),
                      bg='#b68d40',
                      fg='white',
             command = lambda:click_btns(2))
btn2.place(x=85, y= 237)
btn3 = Button(main, text='3', width=4, height= 2,
             font=('PT Heading', 18, 'bold'),
                      bg='#b68d40',
                      fg='white',
             command = lambda:click_btns(3))
btn3.place(x=162, y= 237)
btn4 = Button(main, text='4', width=4, height= 2,
             font=('PT Heading', 18, 'bold'),
                      bg='#b68d40',
                      fg='white',
             command = lambda:click_btns(4))
btn4.place(x=8, y= 151)
btn5 = Button(main, text='5', width=4, height= 2,
             font=('PT Heading', 18, 'bold'),
                      bg='#b68d40',
                      fg='white',
             command = lambda:click_btns(5))
btn5.place(x=85, y= 151)
btn6 = Button(main, text='6', width=4, height= 2,
             font=('PT Heading', 18, 'bold'),
                      bg='#b68d40',
                      fg='white',
             command = lambda:click_btns(6))
btn6.place(x=162, y= 151)
btn7 = Button(main, text='7', width=4, height= 2,
             font=('PT Heading', 18, 'bold'),
                      bg='#b68d40',
                      fg='white',
             command = lambda:click_btns(7))
btn7.place(x=8, y= 65)
btn8 = Button(main, text='8', width=4, height= 2,
             font=('PT Heading', 18, 'bold'),
                      bg='#b68d40',
                      fg='white',
             command = lambda:click_btns(8))
btn8.place(x=85, y= 65)
btn9 = Button(main, text='9', width=4, height= 2,
             font=('PT Heading', 18, 'bold'),
                      bg='#b68d40',
                      fg='white',
             command = lambda:click_btns(9))
btn9.place(x=162, y= 65)
btn_clear = Button(main, text='C', width=4, height= 2,
             font=('PT Heading', 18, 'bold'),
                      bg='#132620',
                      fg='white',
                  command = lambda:clear_btn())
btn_clear.place(x=85, y= 323)
btn_equal = Button(main, text='=', width=4, height= 2,
             font=('PT Heading', 18, 'bold'),
                      bg='#132620',
                      fg='white',
                  command = lambda:resule_btn())
btn_equal.place(x=162, y= 323)
btn_plus = Button(main, text='+', width=4, height= 2,
             font=('PT Heading', 18, 'bold'),
                      bg='#132620',
                      fg='white',
                 command = lambda:click_btns('+'))
btn_plus.place(x=236, y= 323)
btn_min = Button(main, text='-', width=4, height= 2,
             font=('PT Heading', 18, 'bold'),
                      bg='#132620',
                      fg='white',
                command = lambda:click_btns('-'))
btn_min.place(x=236, y= 237)
btn_mult = Button(main, text='*', width=4, height= 2,
             font=('PT Heading', 18, 'bold'),
                      bg='#132620',
                      fg='white',
                 command = lambda:click_btns('*'))
btn_mult.place(x=236, y= 151)
btn_div = Button(main, text='/', width=4, height= 2,
             font=('PT Heading', 18, 'bold'),
                      bg='#132620',
                      fg='white',
                command = lambda:click_btns('/'))
btn_div.place(x=236, y= 65)
# ----------
main.mainloop() # don't remove this  
```
